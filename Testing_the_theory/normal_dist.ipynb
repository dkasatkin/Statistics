```python
# Импортируем библиотеки

import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
# Задаем настройки

from matplotlib import rcParams
rcParams['figure.figsize'] = (16, 10)
sns.set_theme(style="darkgrid")
sns.set_palette('deep')
```

```python
mu, sigma, n = 40, 6, 100000
normal_distribution = np.random.normal(mu, sigma, n)
```

```python
print('var = ', np.var(normal_distribution), '\n',
      'mean = ', np.mean(normal_distribution))
```

```python
sns.histplot(normal_distribution, stat='density')
```

```python
m = 20
random_var_ddof_0 = np.array([])
random_var_ddof_1 = np.array([])


for i in np.arange(m):
    
    random_sample = np.array([])
    
    for _ in np.arange(m):
    
        random_index = np.random.randint(n)
        random_value = normal_distribution[random_index]
        random_sample = np.append(arr=random_sample, values=random_value)

    sample_var_ddof_0 = np.var(random_sample, ddof=0)
    random_var_ddof_0 = np.append(arr=random_var_ddof_0, values=sample_var_ddof_0)
    
    sample_var_ddof_1 = np.var(random_sample, ddof=1)
    random_var_ddof_1 = np.append(arr=random_var_ddof_1, values=sample_var_ddof_1)

one = [sample_var_ddof_0, sample_var_ddof_1]
```
